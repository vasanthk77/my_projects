```python
import pandas as pd
```

```python
a=pd.read_csv('train.csv',error_bad_lines=False,engine='python')
df=pd.DataFrame(a)
df.head()
```

```python
## checking the nan values
df.isnull().sum()
```

```python
## droping nan values
df1=df.dropna()
df1.isnull().sum()
```

```python
## independent and dependent features
X=df1.drop('label',axis=1)
y=df1['label']
```

```python
import tensorflow as tf
```

```python
## importing all modules which was required for lstm
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
## take the vocabulary size from your dataset like taking unique wordss
vocab_size=5000
```

```python
## 3.Data preprocessing using NLP
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import re
nltk.download('punkt')
nltk.download('stopwords')
ps=PorterStemmer()
```

```python
messages=X.copy()
messages.reset_index(inplace=True)
```

```python
## cleaning the data
corpus=[]
for i in range(0,len(messages)):
  new=re.sub('[^a-zA-Z]',' ',messages['title'][i])
  new=new.lower()
  new=new.split()

  new=[ps.stem(word) for word in new if not word in stopwords.words('english')]
  new=' '.join(new)
  corpus.append(new)
```

```python
 corpus
```

```python
## applying one hot encoding for corpus data
onehot_repre=[one_hot(word,vocab_size) for word in corpus]
```

```python
## here we got index values of words based on vocabulary
onehot_repre
```

```python
corpus[10]
```

```python
onehot_repre[10] ## one hot representation for copurs 100th sentence
```

```python
## using padding we can fixed length for every corpus sentence bcoz we make sure that we need to fixed length for every sentence
sent_length=15
embedding_layer=pad_sequences(onehot_repre,padding='post',maxlen=sent_length) ## post add zeros at end of the words
#pre add zeros before words
print(embedding_layer)
```

```python
embedding_layer[1]
```

```python
## Model creation

embedding_vector_features=100
model=Sequential()
model.add(Embedding(vocab_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(200)) ## with 200 neurons
model.add(Dense(1,activation='sigmoid'))  ## output layer
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
## convert embedded_layers list to arrays
import numpy as np
X=np.array(embedding_layer)
y=np.array(y)
```

```python
## train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
## Model Training
model.fit(X_train,y_train,validation_split=0.33,batch_size=100,epochs=10)
```

```python
## Model testing
y_pred=model.predict(X_test)
y_pred=np.where(y_pred>0.5,1,0)
```

```python
## Metrics for the data
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
cm=confusion_matrix(y_test,y_pred)
print(cm)
cr=classification_report(y_test,y_pred)
print(cr)
```
