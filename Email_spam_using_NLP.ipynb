### **Project**:Email spam


```python
import pandas as pd
```

```python
a=pd.read_csv("spam.csv")
```

```python
df=pd.DataFrame(a)
```

```python
df.head()
```

```python
## data cleaning and preprocessing
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stemmer=PorterStemmer()
nltk.download('punkt')
```

```python
corpus1=[]
for i in range(0,len(df['Message'])):
  new=re.sub('[^a-zA-Z]','',df.Message[i])
  new1=new.lower()

  corpus1.append(new1)
```

```python
for i in corpus1:
  words=nltk.sent_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```python
## another way
corpus2=[]
for i in range(0,len(df.Message)):
  new=re.sub('[a-zA-Z]','',df.Message[i])
  new=new.lower()
  new=new.split()
  new=[stemmer.stem(word) for word in new if not word in stopwords.words('english') ]
  new=''.join(new)
  corpus2.append(new)
```

```python
## creating baf of words model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,binary=True)
```

```python
X=cv.fit_transform(corpus1).toarray()
```

```python
## converting ylabel data to 0 and 1
y=pd.get_dummies(df.Message)
y=y.iloc[:,1].values
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
## Model selection
from sklearn.naive_bayes import MultinomialNB
naive=MultinomialNB().fit(X_train,y_train)
```

```python
## testing the model
y_pred=naive.predict(X_test)
```

```python
## Classification metrics
from sklearn.metrics import classification_report,accuracy_score
score=accuracy_score(y_test,y_pred)
print(score)
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
## testing with another model
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred1=model.predict(X_test)
```

```python
## report
score=accuracy_score(y_test,y_pred1)
print(score)
cr=classification_report(y_test,y_pred1)
print(cr)
```
